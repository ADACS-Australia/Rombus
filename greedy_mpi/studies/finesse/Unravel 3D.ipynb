```python

tdat3d = np.load("TemperatureChange_3D_SurfaceHeating_x0Var.npz", allow_pickle=True)
x3 = tdat3d['x']
y3 = tdat3d['y']
z3 = tdat3d['z']
Temp = tdat3d['Temperature'][6]

cm = plt.get_cmap('inferno')
cs = Temp
cNorm = matplotlib.colors.Normalize(vmin=min(cs), vmax=max(cs))
scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=cm)
fig= plt.figure(figsize=(10,6))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x3, y3, z3, c=scalarMap.to_rgba(cs), alpha=0.1)
scalarMap.set_array(cs)
cb=fig.colorbar(scalarMap)
ax.set_xlabel('x [m]')
ax.set_ylabel('y [m]')
ax.set_zlabel('z [m]')
cb.ax.set_ylabel('$\Delta$T [K]')
plt.tight_layout()
```
